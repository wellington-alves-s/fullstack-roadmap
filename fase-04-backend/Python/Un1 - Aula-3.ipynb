```python
#Estruturas de repetição em Python

numeros = [1, 2, 3, 4, 5]
for num in numeros:
    print (num)
```

```python
numero = int(input('Digite um numero (ou 0 para sair):'))

while numero != 0:
    if numero % 2 == 0:
        print('O numero é par.')
    else: 
        print('O numero é impar.')
    numero = int(input('Digite outro numero (ou 0 para sair):'))
```

```python
#Controle de repetição: range, break e continue

#Repetição por quantidade

for x in range (5): #Lembre-se que o range começa do 0
    print(x)
```

```python
#Limite inicial e superior

for y in range (2, 7): #Observe que aqui o 7 não entra no looping
    print(y)
```

```python
#Com incremento

for z in range (1, 11, 2): #Observe que aqui o 11 não entra no looping
    print(z)
```

```python
#Break

for numero in range(2, 11): #Para quando o numero dividido por 2 for igual a zero
    if numero % 2 == 0:
        print('O numero par encontrado é', numero)
        break
```

```python
#Continue

for numero in range(1, 11): #Pula o numero especificado
    if numero == 6:
        continue
    print(numero)
        
```

```python
#Lista de filmes para classificação

filmes = ["Filme 1", "Filme 2", "Filme 3", "Filme 4", "Filme 5"]

#Mensagem de boas vindas
print("Bem-vindo ao sistema de classificação de filmes!")
print("Você tem cinco filmes para classificar.")
print("Digite '0' a qualquer momento para parar.\n")

#Loop para iterar sobre cada filme na lista
for filme in filmes:
    #solicita a classificação do filme ao usuário
    classificacao = input(f"Como você classificaria '{filme}' de 1 a 5? (ou '0' para parar): ")

    #Verifica se o usuário deseja parar
    if classificacao == '0':
        print("Que pena que você não irá classificar mais os filmes.")
        break #Encerra o loop principal

#Converte a classificação para um número inteiro
    classificacao = int(classificacao)

#Verifica se a classificação está dentro do intervalo válido
    if classificacao < 1 or classificacao > 5:
        print("Por favor, digite uma classificacao válida de 1 a 5.")
    else:
        print(f"Você classificou '`{filme}' com a {classificacao} estrelas.\n")

#Mensagem de agradecimento ao finalizar
print ("Obrigado por classificar os filmes!")





```
