```python
Nota_1 = int(input('Digite a primeira nota:'))
Nota_2 = int(input('Digite a segunda nota:'))
Nota_3 = int(input('Digite a terceira nota:'))
Nota_4 = int(input('Digite a quarta nota:'))
#observe que utilizamos a função int(), pois sem ela o Python entende que o valor digitado é uma string
```

```python
#Condição para aprovação do aluno
media = (Nota_1 + Nota_2 + Nota_3 + Nota_4) / 4
if media >= 6:
    print(f'Parabéns {nome}, você foi aprovado com a média {media}!')
else:
    print(f'Que pena {nome}, você foi reprovado com a média {media}. Tente novamente!')
```

```python
#Condição para aprovação do aluno
media = (Nota_1 + Nota_2 + Nota_3 + Nota_4) / 4
if media >= 6:
    situacao = 'aprovado'
else:
    situacao = 'reprovado'
```

```python
#dadas as notas a média final e a situação do aluno
print(f'{nome}, a média das notas é: {media}.')
print(f'Situação do aluno: {situacao}.')
```
