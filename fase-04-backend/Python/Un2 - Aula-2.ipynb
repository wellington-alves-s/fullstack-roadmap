```python
# add(), in e remove()

# Criando um conjunto vazio
meu_conjunto = set()
```

```python
meu_conjunto
```

```python
# Adcionando elementos ao conjunto
meu_conjunto.add(10)
meu_conjunto.add(20)
meu_conjunto.add(30)

# Imprimindo o conjunto
print("Conjunto após adição de elementos:", meu_conjunto)
```

```python
# Verificando a existência de um elemento no conjunto
elemento = 20
if elemento in meu_conjunto:
    print(f"O elemento {elemento} está presente no conjunto.")
else:
    print(f"O elemento {elemento} não está presente no conjunto.")

# Removendo um elemento do conjunto
meu_conjunto.remove(20)

# Imprimindo o conjunto após a remoção
print("Conjunto após remoção do elemento 20:", meu_conjunto)
```

```python
# Objetos do tipo mapping: dicionários

# Exemplo 1 - Criando um dicionário vazi, seguido de adição de pares chave-valor
dici_1 = {}
dici_1["nome"] = "Maria"
dici_1["idade"] = 25



```

```python
dici_1
```

```python
# Exemplo 2 - Criando um dicionário com pares chave-valor
dici_2 = {"nome": "Maria", "idade": 25}

# Exemplo 3 - Criação de um dicionario com uma lista de tuplas representando pares chave-valor
dici_3 = dict([("nome", "Maria"), ("idade", 25)])

# Exemplo 4 - Criação de um dicionario usando a função built-in zip() e duas listas, uma para chaves e outra para valores
dici_4 = dict(zip(["nome", "idade"], ["Maria", 25]))

# Teste de se todas as construções resultam em objetos iguais
print(dici_1 == dici_2 == dici_3 == dici_4)  # Deve retornar True
print(dici_1)
```

```python
# Objetos do tipo array Numpy

# Importando a biblioteca Numpy para manipulação de arrays
import numpy as np

# Crie um array Numpy de numeros inteiros
my_array = np.array([1, 2, 3, 4, 5])

# Imprima o array e seu tipo
print("Array original:")
print(my_array)

```

```python
# Realize operações matemáticas com o array
squared_array = my_array ** 2 # Eleva cada elemento ao quadrado
sum_of_elements = np.sum(my_array) # Soma todos os elementos do array

# Imprima os resultados das operações
print("\nArray ao quadrado:")
print(squared_array)
print("\nSoma dos elementos do array:")
print(sum_of_elements)

# Acesse elementos do array
element_at_index_2 = my_array[2] # Acessa o terceiro elemento (índice 2)
print("\nElemento no índice 2:", element_at_index_2)
```

```python
# Importe as bibliotecas necessárias
import numpy as np

# Dados dos participantes
participantes = [
{
    "nome": "Alice", 
    "localização": "EUA",
    "afiliacao": "Universidade A",
    "interesses": ["Fisica", "Astronomia"]
},
{
    "nome": "Bob",
    "localização": "Brasil",
    "afiliacao": "Instituto B",
    "interesses": ["Biologia", "Astronomia"]
},
{
    "nome": "Charlie",
    "localização": "India",
    "afiliacao": "Universidade C",
    "interesses": ["Quimica", "Engenharia"]
}
# Adicione mais participantes conforme necessário
]

# Usando sets para identificar diferentes regiões dos participantes
regioes = set(participante["localização"] for participante in participantes)

# Usando um dicionario para categorizar afiliações
afiliacoes = {}
for participante in participantes:
    afiliacao = participante["afiliacao"]
    if afiliacao not in afiliacoes:
        afiliacoes[afiliacao] = []
    afiliacoes[afiliacao].append(participante["nome"])

# Usando Numpy arrays para analisar areas de interesses
areas_de_interesse = np.array([interesse for participante in participantes for interesse in participante["interesses"]])
interesse_unicos, contagem = np.unique(areas_de_interesse, return_counts=True)
area_mais_popular = interesse_unicos[np.argmax(contagem)]

# Resultados
print("Regiões dos participantes:", regioes)
print("Afiliacões dos participantes:",)
for afiliacao, nomes in afiliacoes.items():
    print(f"  {afiliacao}: {nomes}")
print("Área de interesse mais popular:", area_mais_popular)
```
